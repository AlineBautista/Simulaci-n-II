```python

import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return np.cos(np.pi * x / 2)

a = 0
b = 1
N = 100000


x_random = np.random.uniform(a, b, N)

f_random = f(x_random)

integral = (b - a) * np.mean(f_random)

x = np.arange(0,1,0.1)
y = np.cos(np.pi * x / 2)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Función f(x) = cos(pi*x/2)')
plt.show()

print(f"El valor aproximado de la integral es: {integral:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función que queremos integrar
def f(x):
    return np.cos(np.pi * x / 2)

# Parámetros para el método de Montecarlo
a = 0       # Límite inferior
b = 1       # Límite superior
N_max = 100000  # Número máximo de puntos aleatorios
step = 100     # Incremento de puntos en cada iteración

# Listas para almacenar el número de puntos y las aproximaciones de la integral
iterations = []
approximations = []

# Calcular la integral utilizando diferentes números de puntos aleatorios
for N in range(step, N_max + 1, step):
    x_random = np.random.uniform(a, b, N)
    f_random = f(x_random)
    integral = (b - a) * np.mean(f_random)

    # Almacenar los resultados
    iterations.append(N)
    approximations.append(integral)

# Valor exacto de la integral
exact_value = 2 / np.pi

# Graficar la convergencia
plt.figure(figsize=(10, 6))
plt.plot(iterations, approximations, label="Aproximación de Montecarlo")
plt.axhline(y=exact_value, color='r', linestyle='--', label="Valor Exacto")
plt.xlabel("Número de puntos aleatorios (N)")
plt.ylabel("Valor de la integral aproximada")
plt.title("Convergencia del Método de Montecarlo")
plt.legend()
plt.grid(True)
plt.show()
```
